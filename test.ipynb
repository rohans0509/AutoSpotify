```python
import spotipy
from spotipy.oauth2 import SpotifyOAuth
from dotenv import load_dotenv
import streamlit as st
import pandas as pd

load_dotenv()
```

```python
scope='user-library-read user-modify-playback-state streaming'
auth_manager = SpotifyOAuth(scope=scope)
sp = spotipy.Spotify(auth_manager=auth_manager)
```

```python
# Make streamlit sight
st.title('Spotify Audio Features')

# Get audio features of last 10 songs
results = sp.current_user_saved_tracks(limit=10)
tracks = results['items']
track_ids = [track['track']['id'] for track in tracks]
audio_features = sp.audio_features(track_ids)

# Display audio features
df= pd.DataFrame(audio_features)
# Add track names and ids, change index
df['track_name'] = [track['track']['name'] for track in tracks]
df['track_id'] = [track['track']['id'] for track in tracks]
df.set_index('track_name', inplace=True)
display(df)
```

```python
# Give recommendations based on a chosen track. Take first row for testing
test_track=tracks[0]
print(test_track['track']['name'], test_track['track']['artists'][0]['name'])
# Get recommendations
recommendations = sp.recommendations(seed_tracks=[test_track['track']['id']])
rec_tracks = recommendations['tracks']
rec_track_names = [track['name'] for track in rec_tracks]
rec_artist_names = [track['artists'][0]['name'] for track in rec_tracks]
for i in range(len(rec_tracks)):
    print(f'{rec_track_names[i]} by {rec_artist_names[i]}')
```

```python
from langchain_google_genai import ChatGoogleGenerativeAI
from langchain_core.messages import HumanMessage, ToolMessage
from langchain_core.tools import tool
from dotenv import load_dotenv
load_dotenv()

@tool
def add(a: int, b: int):
    """Adds a and b.

    Args:
        a: first int
        b: second int
    """
    return a + b


@tool
def multiply(a: int, b: int):
    """Multiplies a and b.

    Args:
        a: first int
        b: second int
    """
    return a * b


tools = [add, multiply]

llm = ChatGoogleGenerativeAI(model="gemini-pro")
llm=llm.bind_tools(tools)
query = "What is 11 + 49? Give number as answer"

# llm.invoke(query).tool_calls
# print(llm.invoke(query).tool_calls)
# print(llm.invoke(query))


messages = [HumanMessage(query)]
ai_msg = llm.invoke(messages)
messages.append(ai_msg)

for tool_call in ai_msg.tool_calls:
    selected_tool = {"add": add, "multiply": multiply}[tool_call["name"].lower()]
    tool_output = selected_tool.invoke(tool_call["args"])
    messages.append(ToolMessage(tool_output, tool_call_id=tool_call["id"]))

llm.invoke(messages)
```
